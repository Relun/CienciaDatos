3. Escribir una funciÃ³n que reciba un diccionario con las notas de los
alumnos en curso en un examen y devuelva una serie con las notas de
los alumnos aprobados ordenadas de mayor a menor.

```python
def notas(nota):
  nota = pd.Series(nota) #convierte el diccionario en serie
  return nota[nota>=4].sort_values(ascending=False) #alumno aprueba con nota 4, ordene la serie de mayor a menor tomando las notas >= a 4

nota = {'Alumno 1': 7, 'Alumno 2': 4, 'Alumno 3': 5, 'Alumno 4': 2} #diccionario con las notas
print(notas(nota))
```
