Para una tarde de bingo se usarán tablas generadas con las siguientes
reglas:

• Los 5 números en cada columna son aleatorios ordenados no
repetidos de acuerdo al rango asignado por columna.

• La casilla central de la tabla tiene valor marcado cero o ‘libre’.
Realice un programa para generar una tabla de bingo

image.png

```python
import pandas as pd
import random

bingo = pd.DataFrame(columns=['B','I','N','G','O']) #creación del dataframe
#con ciclo for se llena cada casilla según el rango indicado y random permite obtener un número al azar
for i, number in [('B', range(1, 16)), ('I', range(16, 31)), ('N', range(31, 46)), ('G', range(46, 61)), ('O', range(61, 76))]:
  bingo_numbers = random.sample(number, 5)
  if i == 'N': #esto llena la casilla central
    bingo_numbers[2] = 'libre'

  bingo[i] = bingo_numbers

print(bingo)
```
