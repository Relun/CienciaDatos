2. El archivo “M3E4 - titanic.csv” contiene información sobre los

```python
import pandas as pd
```

a. Generar un DataFrame con los datos del fichero.

```python
titanic = 'Titanic.csv'

df = pd.read_csv(titanic, index_col = 0)
print(df)
```

b. Mostrar por pantalla las dimensiones del DataFrame, el número
de datos que contiene, los nombres de sus columnas y filas, los
tipos de datos de las columnas, las 10 primeras filas y las 10
últimas filas.

```python
print("1. Dimensiones ", df.ndim)
print("2. Número de datos ", df.size)
#print("Nombres columnas ", df.columns)
print("3. Nombres columnas ", df.columns.values)
print("4. Nombres filas ", df.index.values)
print("5. Tipos de datos de las columnas ", df.dtypes)
print("6. 10 primeras filas ", df.head(10))
print("7. 10 últimas filas ", df.tail(10))

```

c. Mostrar por pantalla los datos del pasajero con identificador

```python
display(df.loc[148])
```

d. Mostrar por pantalla las filas pares del DataFrame.

e. Mostrar por pantalla los nombres de las personas que iban en
primera clase ordenadas alfabéticamente.

```python
display(df[df['Pclass'] == 1]['Name'].sort_values()) #Pclass (clase 1, 2 y 3)
```

f. Mostrar por pantalla el porcentaje de personas que
sobrevivieron y murieron.

```python
#Survived (0 = no sobrevivio; 1= sí sobrevivio)
#survivors = 100 * df.Survived.sum() / len(df)
survivors = 100 * df['Survived'].sum() / len(df)
print(f'Sobrevivientes: {survivors}%, Muertos: {100-survivors}%')

```

g. Mostrar por pantalla el porcentaje de personas que
sobrevivieron en cada clase.


```python
display(df.groupby('Pclass')['Survived'].sum() / len(df) * 100) #Con respecto al 38.38% de los sobrevivientes
```

```python
df_no_age = df.dropna(how='any', axis=0, subset='Age', inplace=False)
display(df_no_age.head(20))
```
