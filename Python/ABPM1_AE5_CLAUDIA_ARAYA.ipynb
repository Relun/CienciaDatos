2. Con el ejercicio anterior, cree un programa que le pida ingresar los
valores al usuario.

```python
#M√©todo dividir, solicita los nros al usuario y hace la divisi√≥n,
#si el nro divisor es 0 manda mensaje de que no debe ser cero
def dividir(dividendo, divisor):
  try:
    resultado = dividendo / divisor
    return print(f"El resultado de la divisi√≥n es: {resultado}")
  except ZeroDivisionError:
    print("El divisor no puede ser cero")
    #raise ZeroDivisionError("El divisor no puede ser cero")

#
ingresar_dividendo = int(input("Ingrese el dividendo "))
ingresar_divisor = int(input("Ingrese el divisor "))
resultado = dividir(ingresar_dividendo, ingresar_divisor)

```

3. Cree una funci√≥n de inter√©s simple, entregando una excepci√≥n si la
tasa es superior a 100.


*   El programa debe recibir: monto, a√±o y tasa.
*   El c√°lculo de inter√©s ser√°: (monto * a√±o * tasa) / 100
*   Agregue mensajes adecuados al contexto, para la interacci√≥n
del programa.







```python
#M√©todo sacar inter√©s, si la tasa es menor a 100 realiza el c√°lculo
#si la tasa es mayor a 100 env√≠a mensaje de que ingrese un nro menor y no realiza el c√°lculo
def interes_simple(monto, anio, tasa):
  try:
    if tasa > 100:
      raise ValueError(tasa)
  except ValueError:
    print(f"La tasa de {tasa} es superior a 100, ingresa un nro menor")
  else:
    resultado = (monto * anio * tasa) / 100
    print(resultado)
    #print(tasa, "La tasa es menor a 100")

ingresar_monto = int(input("Ingrese el monto "))
ingresar_anio = int(input("Ingrese los a√±os "))
ingresar_tasa = int(input("Ingrese la tasa "))
resultado = interes_simple(ingresar_monto, ingresar_anio, ingresar_tasa)
```
